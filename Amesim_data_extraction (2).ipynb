```python
df3=pd.read_excel('/content/output_simulated_potential_2.xlsx')
df3.head()
```

```python
def clean_time(x):
  y=x.strip().split(' ')
  return float(y[0])

def clean_potential(x):
  y=x.strip().split(' ')
  return float(y[-1])
```

```python
df3['Time']=df3[df3.columns[0]].apply(clean_time)
df3['Potential']=df3[df3.columns[0]].apply(clean_potential)
simulated_data=df3[['Time','Potential']]
```

```python
simulated_data
```

```python
simulated_data=simulated_data[(994.959<df3['Time']) & (simulated_data['Time']<2427.104)]
simulated_data.to_csv('simulated_data.csv')
```

```python
fig2=px.line(data_frame=simulated_data,y='Potential',x='Time')
fig1=px.line(data_frame=df1,y=['B_Cell_Voltage_N','B_Cell_Voltage_N_plus_1',\
            'B_Cell_Voltage_N_plus_2','B_Cell_Voltage_N_plus_3','B_Cell_Voltage_N_plus_4'],x='Time')
fig = make_subplots(rows=2, cols=1)
fig.add_trace(fig1.data[0],row=1, col=1)
fig.add_trace(fig1.data[1],row=1, col=1)
fig.add_trace(fig1.data[2],row=1, col=1)
fig.add_trace(fig1.data[3],row=1, col=1)
fig.add_trace(fig1.data[4],row=1, col=1)
fig.add_trace(fig2.data[0],row=2, col=1)
```
